## Preprocess Data for CVAE

The aim of this notebook is to translate NetCDF files (.nc) of three daily climate variables (maximum temperature, precipitations, wind) to a numpy 3D-array. This output array can easily be read for training and evaluating the Convolutional Variational AutoEncoder model.

#### 0. Libraries

Import the following libraries to: 
- handle data arrays (numpy and xarray), 
- display results on charts and maps (matplotlib and cartopy), 
- handle dates (datetime), 
- create and save data frames (csv and pandas).

```python
import numpy as np
import xarray as xr
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
from datetime import datetime
import csv
import pandas as pd
```

#### 1. Load Data to xarray

We load both historical and projection files (here 1975-1999 and 2065-2089) for the three daily variables: maximum temperature (tasmax), precipitations (pr), and wind speed (sfcWind).

```python
temp_histo = xr.open_dataset("../data/tasmax_day_CMCC-ESM2_historical_r1i1p1f1_gn_19750101-19991231.nc")
temp_proj = xr.open_dataset("../data/tasmax_day_CMCC-ESM2_ssp585_r1i1p1f1_gn_20650101-20891231.nc")
prcp_histo = xr.open_dataset("../data/pr_day_CMCC-ESM2_historical_r1i1p1f1_gn_19750101-19991231.nc")
prcp_proj = xr.open_dataset("../data/pr_day_CMCC-ESM2_ssp585_r1i1p1f1_gn_20650101-20891231.nc")
wind_histo = xr.open_dataset("../data/sfcWind_day_CMCC-ESM2_historical_r1i1p1f1_gn_19750101-19991231.nc")
wind_proj = xr.open_dataset("../data/sfcWind_day_CMCC-ESM2_ssp585_r1i1p1f1_gn_20650101-20891231.nc")
```

The Greenwich meridian (at longitude 0) goes through Europe. In the 0-360 data, the continent is split into two parts: one after 0 and one before 360. We therefore shift the data from -180 to 180 to avoid further problems in the process.

```python
# The original data goes from longitudes 0 to 360
# Here we reorder them from -180 to 180
# to handle Europe (including Greenwich meridian at longitude 0) 

temp_histo.coords['lon'] = (temp_histo.coords['lon'] + 180) % 360 - 180
temp_histo = temp_histo.sortby(temp_histo.lon)
temp_proj.coords['lon'] = (temp_proj.coords['lon'] + 180) % 360 - 180
temp_proj = temp_proj.sortby(temp_proj.lon)

prcp_histo.coords['lon'] = (prcp_histo.coords['lon'] + 180) % 360 - 180
prcp_histo = prcp_histo.sortby(prcp_histo.lon)
prcp_proj.coords['lon'] = (prcp_proj.coords['lon'] + 180) % 360 - 180
prcp_proj = prcp_proj.sortby(prcp_proj.lon)

wind_histo.coords['lon'] = (wind_histo.coords['lon'] + 180) % 360 - 180
wind_histo = wind_histo.sortby(wind_histo.lon)
wind_proj.coords['lon'] = (wind_proj.coords['lon'] + 180) % 360 - 180
wind_proj = wind_proj.sortby(wind_proj.lon)
```

#### 2. Restrict to Western Europe (Square)

The neural network takes square "images" as inputs: we select a square geographical data frame over Western Europe.

```python
# select relevant latitude and longitude
# depending on the desired size and location

# 32*32 square for the neural network
min_lon = -10
max_lon = 29
min_lat = 36
max_lat = 66
# (-10, 68.5, 36, 71) for Europe
# (-10, 35, 36, 71) for 37x37 square
# (-10, 24, 36, 62) for 28x28 square
```

```python
# slice of the datasets accordingly

sq_temp_histo = temp_histo.sel(lon = slice(min_lon, max_lon),
                      lat = slice(min_lat, max_lat))
sq_temp_proj = temp_proj.sel(lon = slice(min_lon, max_lon),
                    lat = slice(min_lat, max_lat))

sq_prcp_histo = prcp_histo.sel(lon = slice(min_lon, max_lon),
                      lat = slice(min_lat, max_lat))
sq_prcp_proj = prcp_proj.sel(lon = slice(min_lon, max_lon),
                    lat = slice(min_lat, max_lat))

sq_wind_histo = wind_histo.sel(lon = slice(min_lon, max_lon),
                      lat = slice(min_lat, max_lat))
sq_wind_proj = wind_proj.sel(lon = slice(min_lon, max_lon),
                    lat = slice(min_lat, max_lat))
```

#### 3. Convert to ndarray Images

The xarray data is converted to numpy arrays, just as if they were images encoded with numbers.

```python
# keep track of the coordinates and time indices
time_list = np.array(sq_temp_histo['time'])
time_proj = np.array(sq_temp_proj['time'])
lat_list = sq_temp_histo.coords['lat']
lon_list = sq_temp_histo.coords['lon']
n_t = len(sq_temp_histo['time'])
n_t_proj = len(sq_temp_proj['time'])
n_lat = len(lat_list)
n_lon = len(lon_list)
```

```python
# create a numpy array for each variable and time period

# dtype is set to float32 to match the data types needed in the CVAE
sq_temp_images = np.ndarray((n_t, n_lat, n_lon, 1), dtype="float32")
sq_prcp_images = np.ndarray((n_t, n_lat, n_lon, 1), dtype="float32")
sq_wind_images = np.ndarray((n_t, n_lat, n_lon, 1), dtype="float32")
sq_proj_temp_images = np.ndarray((n_t_proj, n_lat, n_lon, 1), dtype="float32")
sq_proj_prcp_images = np.ndarray((n_t_proj, n_lat, n_lon, 1), dtype="float32")
sq_proj_wind_images = np.ndarray((n_t_proj, n_lat, n_lon, 1), dtype="float32")

for t in range(n_t):
    sq_temp_images[t, :, :, 0] = sq_temp_histo['tasmax'][t, :, :]
    sq_prcp_images[t, :, :, 0] = sq_prcp_histo['pr'][t, :, :]
    sq_wind_images[t, :, :, 0] = sq_wind_histo['sfcWind'][t, :, :]
for t in range(n_t_proj):
    sq_proj_temp_images[t, :, :, 0] = sq_temp_proj['tasmax'][t, :, :]
    sq_proj_prcp_images[t, :, :, 0] = sq_prcp_proj['pr'][t, :, :]
    sq_proj_wind_images[t, :, :, 0] = sq_wind_proj['sfcWind'][t, :, :]
```

```python
# flip images so that they are in the right North-South orientation
sq_temp_images = np.flip(sq_temp_images, axis=1)
sq_prcp_images = np.flip(sq_prcp_images, axis=1)
sq_wind_images = np.flip(sq_wind_images, axis=1)
sq_proj_temp_images = np.flip(sq_proj_temp_images, axis=1)
sq_proj_prcp_images = np.flip(sq_proj_prcp_images, axis=1)
sq_proj_wind_images = np.flip(sq_proj_wind_images, axis=1)
```

#### 4. Normalize

We normalize each variable individually, taking the extrema over both historical and projection data. That is, for a given climate variable $x$, each data point $x_i$ will be updated as follows:
$$
x_i := \frac{x_i - min_{j \in histo, proj}(x_j)}{max_{j \in histo, proj}(x_j) - min_{j \in histo, proj}(x_j)}
$$

To avoid extra computation time, we directly input the right values for the periods 1975-1999 and 2065-2089, but the extrema have to be re-run when the notebook is used out of this scheme.

```python
#max_tasmax = float(np.max(proj)['tasmax'])
# on Europe train data: 0.001352, 12/02/1994
# on world projection data: 0.008302, 01/01/2090
max_tasmax = 332.86590576171875
#max_tasmax = float(np.max(proj)['tasmax'])
#min_tasmax = float(np.min(histo)['tasmax'])
min_tasmax = 174.30337524414062
```

```python
#max_pr = float(np.max(prcp_proj)['pr'])
#min_pr = float(np.min(prcp_proj)['pr'])
max_pr = 0.008302416652441025
min_pr = 0
```

```python
#max_wind = float(np.max(wind_proj)['sfcWind'])
#min_wind = float(np.min(wind_proj)['sfcWind'])
max_wind = 34.41741180419922
min_wind = 0.023357095196843147
```

```python
# create new normalized arrays
norm_temp_images = (sq_temp_images-min_tasmax)/(max_tasmax-min_tasmax)
norm_proj_temp_images = (sq_proj_temp_images-min_tasmax)/(max_tasmax-min_tasmax)
norm_prcp_images = (sq_prcp_images-min_pr)/(max_pr-min_pr)
norm_proj_prcp_images = (sq_proj_prcp_images-min_pr)/(max_pr-min_pr)
norm_wind_images = (sq_wind_images-min_wind)/(max_wind-min_wind)
norm_proj_wind_images = (sq_proj_wind_images-min_wind)/(max_wind-min_wind)
```

```python
x_lat = 0
x_lon = 0
plt.plot(norm_temp_images[:,x_lat,x_lon,0], 'b', label='history')
plt.plot(norm_proj_temp_images[:,x_lat,x_lon,0], 
         'r', label='projection')
plt.title('Temperature evolution and comparison on two periods \nat the point of coordinates ('
          +str(lat_list[x_lat])[:5]+','+str(lon_list[x_lon])[:5]+')')
plt.xlabel('Time (days)')
plt.ylabel('Normalized Temperature')
plt.legend()

plt.show()

print('There is a', 
      round(np.average(sq_proj_temp_images)-np.average(sq_temp_images), 2), 
      'degree difference between the averages of each time series.')
```

#### 5. Split into Train & Test Data

The CVAE will learn and train on historical data, as a calibration on a "normal" situation. It is split into a train data set (most of the data, e.g. 80%) and a test data set, on which the CVAE performance will be monitored.

```python
train_proportion = 0.8
train_t = int(train_proportion*n_t)
test_t = n_t - train_t
```

```python
train_temp_images = norm_temp_images[0:train_t,:,:,:]
train_prcp_images = norm_prcp_images[0:train_t,:,:,:]
train_wind_images = norm_wind_images[0:train_t,:,:,:]
test_temp_images = norm_temp_images[train_t:,:,:,:]
test_prcp_images = norm_prcp_images[train_t:,:,:,:]
test_wind_images = norm_wind_images[train_t:,:,:,:]
```

#### 6. Visualize

```python
height = n_lat/10
width = n_lon/10

fig = plt.figure(figsize=(width*3, height))
fig.suptitle('Average Temperature in Europe', fontsize=16)

ax_histo = fig.add_subplot(1, 2, 1, projection=ccrs.PlateCarree())
ax_histo.set_title('History (1975-1999)')
cs = ax_histo.imshow(np.mean(norm_temp_images[:,:,:,:], axis=0), cmap='hot_r', 
          extent=[lon_list[0], lon_list[-1], lat_list[0], lat_list[-1]], 
                     vmin=0.6, vmax=0.9) # fix the colorbar min and max
cbar = plt.colorbar(cs)
ax_histo.coastlines()

ax_proj = fig.add_subplot(1, 2, 2, projection=ccrs.PlateCarree())
ax_proj.set_title('SSP5-8.5 Projections (2065-2089)')
cs_norm = ax_proj.imshow(np.mean(norm_proj_temp_images[:,:,:,:], axis=0), cmap='hot_r', 
          extent=[lon_list[0], lon_list[-1], lat_list[0], lat_list[-1]], 
                         vmin=0.6, vmax=0.9)
cbar_norm = plt.colorbar(cs_norm)
ax_proj.coastlines()

plt.show()
```

####  7. Make a 3D-Image

Now the three arrays are assembled into one 3-dimension array - one for each time period.

```python
norm_images = np.zeros((n_t, n_lat, n_lon, 3), dtype="float32")
norm_images[:,:,:,0] = norm_temp_images.reshape(n_t,n_lat,n_lon)
norm_images[:,:,:,1] = norm_prcp_images.reshape(n_t,n_lat,n_lon)
norm_images[:,:,:,2] = norm_wind_images.reshape(n_t,n_lat,n_lon)

norm_proj_images = np.zeros((n_t_proj, n_lat, n_lon, 3), dtype="float32")
norm_proj_images[:,:,:,0] = norm_proj_temp_images.reshape(n_t_proj,n_lat,n_lon)
norm_proj_images[:,:,:,1] = norm_proj_prcp_images.reshape(n_t_proj,n_lat,n_lon)
norm_proj_images[:,:,:,2] = norm_proj_wind_images.reshape(n_t_proj,n_lat,n_lon)
```

####  8. Write Data

```python
# convert the time list to easily readable dates
time_list = sq_temp_histo['time']
time_list = time_list.indexes['time'].to_datetimeindex()
# then to a pandas data frame to be saved as a csv file
time_list = pd.DataFrame(time_list)
time_list.to_csv('input/dates_histo_data.csv')
```

```python
# create and save the train and test time lists
time_list_train = time_list[:7300]
time_list_test = time_list[7300:]
time_list_train.to_csv('input/dates_train_data.csv')
time_list_test.to_csv('input/dates_test_data.csv')
```

```python
# save projection time list
time_proj = sq_temp_proj['time'].indexes['time'].to_datetimeindex()
pd.DataFrame(time_proj).to_csv('input/dates_proj_data.csv')
```

```python
# save 3d-datasets
np.save("../input/preprocessed_3d_histo_data.npy", norm_images)
np.save("../input/preprocessed_3d_proj_data.npy", norm_proj_images)
```

```python
# save train and test 3d-datasets separately
np.save("../input/preprocessed_3d_train_data.npy", norm_images[:int(train_proportion*n_t),:,:,:])
np.save("../input/preprocessed_3d_test_data.npy", norm_images[int(train_proportion*n_t):,:,:,:])
```
